```python
import gradio as gr

def chat_function(user_input, history):
    if history:
        last_user, last_bot = history[-1]
        response = (
            f"You asked '{user_input}'. "
            f"Previously you said '{last_user}'. "
            "I am continuing the conversation."
        )
    else:
        response = "Hello! How can I help you today?"

    return response


chatbot_ui = gr.ChatInterface(
    fn=chat_function,
    title="Milestone 3 Chatbot",
    description="Conversational Interface with Dialogue Management"
)
chatbot_ui.launch()


```
